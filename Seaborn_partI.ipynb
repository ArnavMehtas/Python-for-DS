# Seaborn 

## Distribution plots
- distplot
- joinplot
- pairplot

```python
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

## Correlation Heatmap

```python
# Check the data types of each column
print(df.dtypes)
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])
# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
correlation_matrix
```

```python
sns.heatmap(correlation_matrix)
```

## JOIN PLOT

#### A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

### Univariate Analysis


```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.jointplot(x='tip',y='size',data=df,kind='reg')
```

```python
sns.jointplot(x='tip',y='sex',data=df,kind='reg')
```

## Pair plot


#### A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Dist plot


#### Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots


```python
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(x='day',data=df)
```

# Bar Plot

```python
sns.barplot(y='smoker',x='sex',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

## Box plot

```python
sns.boxplot(y='total_bill',x='smoker',data=df)
```

```python
sns.barplot(x='day',y='sex',data=df,palette='rainbow')
```

## Violin plot

```python
sns.violinplot(y='day',x='total_bill',data=df, palette='rainbow')
```
